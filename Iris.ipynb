Aim : To Develop a Model that can classify iris flowers into differrent species based on their sepal and petal measurment.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv("/content/drive/MyDrive/Python_dataset/IRIS.csv")
df = pd.DataFrame(data)
df
```

```python
df.loc[df['species'] == 'Iris-setosa', 'species'] = 0
df.loc[df['species'] == 'Iris-versicolor', 'species'] = 1
df.loc[df['species'] == 'Iris-virginica', 'species'] = 2

df['species'] = df['species'].astype(int)
```

# Model Training :   

```python
x = df[["sepal_length","sepal_width","petal_length","petal_width"]]
y = df["species"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
Model = LogisticRegression()
Model.fit(x_train,y_train)
```

```python
# Evaluating the model to verify the accuray

print("Accuracy : ",Model.score(x_test,y_test)*100)
print("Accuracy : ",Model.score(x_train,y_train)*100)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Building machine learning model and training them
Model2 = KNeighborsClassifier()
Model2.fit(x_train,y_train)
```

```python
# Evaluating the model to verify the accuray

print("Accuracy : ",Model2.score(x_test,y_test)*100)
print("Accuracy : ",Model2.score(x_train,y_train)*100)
```
